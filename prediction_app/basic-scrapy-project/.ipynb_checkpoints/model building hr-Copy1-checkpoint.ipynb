```python
import pandas as pd
import numpy as np
import openpyxl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager
import pickle
from matplotlib.legend_handler import HandlerLine2D
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, auc,roc_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV, PassiveAggressiveClassifier, RidgeClassifierCV, SGDClassifier, Perceptron
from sklearn.naive_bayes import BernoulliNB, GaussianNB
from xgboost import XGBClassifier
#import shap
```

```python
df_final_all = pd.read_excel("final output all profile 1.xlsx")
```

```python
df_final_all.dropna(subset=["number_of_jobs"],inplace=True)
df_final_all
```

```python
df_final_all = df_final_all.fillna(0)
```

```python
df_final_all.isnull().sum().sum()
```

```python
df_final_all['job_change'].value_counts(normalize=True)
```

```python
seednumber = 10
shuffle_param = True
np.random.seed(seednumber)

X = df_final_all.drop(["Unnamed: 0","job_change","job_change_horizon","profile_name"], axis=1)
y = df_final_all['job_change']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle=shuffle_param)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =0)
MLA_columns = []
MLA_compare = pd.DataFrame(columns = MLA_columns)

MLA = [
    #KNN
    KNeighborsClassifier(3),
    #Trees
    DecisionTreeClassifier(),
    #ensemble
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    BaggingClassifier(),
    ExtraTreesClassifier(),
    XGBClassifier(),
    #Navie Bayes
    BernoulliNB(),
    GaussianNB(),
    #GLM
    LogisticRegression(),
    LogisticRegressionCV(),
    PassiveAggressiveClassifier(),
    RidgeClassifierCV(),
    SGDClassifier(),
    Perceptron()   
    ]
```

```python
row_index = 0
for alg in MLA:
    alg.fit(X_train, y_train)
    y_pred = alg.predict(X_test)
    MLA_name = alg.__class__.__name__
    MLA_compare.loc[row_index,'MLA Name'] = MLA_name
    MLA_compare.loc[row_index,'Accuracy'] = accuracy_score(y_test, y_pred)
    #MLA_compare.loc[row_index,'MLA Precission'] = precision_score(y_test, y_pred)
    #MLA_compare.loc[row_index,'MLA Recall'] = recall_score(y_test, y_pred)
    row_index+=1

MLA_compare.sort_values(by = ['Accuracy'], ascending = False, inplace = True) 

plt.subplots(figsize=(8,6))
sns.barplot(x = "MLA Name", y = "Accuracy", data = MLA_compare, palette=("Blues_d"))
plt.xticks(rotation=90)
plt.title('MLA Accuracy Comparison')
plt.show()

MLA_compare['Accuracy'] = MLA_compare['Accuracy'].apply(lambda x: "{0:.2f}%".format(x*100))

print(MLA_compare)
```

```python
gb_clf2 = XGBClassifier()
#learning_rate = 0.1, n_estimators = 16, max_depth = 1, min_samples_split = 0.3,  min_samples_leaf = 0.1
gb_clf2.fit(X_train, y_train)
```

```python
y_pred = gb_clf2.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

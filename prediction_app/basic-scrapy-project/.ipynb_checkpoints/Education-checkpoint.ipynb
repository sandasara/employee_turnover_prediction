```python
import ast
import re

# Define a dictionary to map course categories to numeric values
course_category_mapping = {
    'Other': 1,
    'High school': 2,
    'Diploma': 3,
    'Bachelor': 4,
    'Masters': 5,
    'Post Masters': 6  # Assuming this category for courses like Ph.D.
}

# Define keyword lists for each category
course_keywords = {
    'Other': [],
    'High school': ['school', 'college', 'convent', 'vidyalaya','Advance Level','GCE','Ordinary level','GCE Advance Level Combined Mathematics'],
    'Diploma': ['diploma', 'dip'],
    'Bachelor': ['bachelor', 'bsc', 'bcom', 'beng','B.Sc.'],
    'Masters': ['masters', 'msc', 'mba','master'],
    'Post Masters': ['phd', 'pfor', 'mphil']
}

# Function to categorize a course based on its details
def categorize_course(course_details):
    course_details = course_details.lower()
    for category, keywords in course_keywords.items():
        for keyword in keywords:
            if re.search(r'\b' + re.escape(keyword) + r'\b', course_details):
                return category
    return 'Other'  # Default to 'Other' if no category is found


# Apply the categorize_course function to create the 'Course Category' column
extracted_education_df['Course Category'] = extracted_education_df['Course Details'].apply(categorize_course)

# Map the course categories to numeric values
extracted_education_df['Course Category'] = extracted_education_df['Course Category'].map(course_category_mapping)

# Print the updated DataFrame
print(extracted_education_df)
```

```python
#country mapping
from geopy.geocoders import Nominatim

def get_country_from_city(city_name):
    geolocator = Nominatim(user_agent="city-to-country")
    location = geolocator.geocode(city_name)

    if location:
        country_name = location.address.split(",")[-1].strip()
        if country_name == "ශ්‍රී ලංකාව இலங்கை":
            country_name = "Sri Lanka"
        return country_name
    else:
        return None


for index, row in df.iterrows():
    city = row['headquaters']  
    country = get_country_from_city(city)

    if country:
        print(f" {country}")
    else:
        print(f"0")
```

```python
df['headquaters'] = df['headquaters'].apply(lambda city: get_country_from_city(city))
```

```python
from datetime import datetime

current_date = datetime(2023, 1, 31)

# Replace 'present' with the current year
df['End Time'] = df['End Time'].replace('present', current_date.year)

# Replace empty strings and NaN values with zeros
df['Start Time'] = df['Start Time'].replace('', '0')
df['End Time'] = df['End Time'].replace('', '0')

# Encoding the columns as per your specifications
current_year = 2023
df['start_recency_months'] = df['Start Time'].astype(int).apply(lambda x: (current_year - x) * 12)
df['end_recency_months'] = df['End Time'].astype(int).apply(lambda x: (current_year - x) * 12)

size_mapping = {
    '1,000 - employees': 1,
    '1,001-5,000 employees': 2,
    '5,001-10,000 employees': 3,
    '10,001+ employees': 4
}

# Replace NaN values with a default value, in this case, 0
df['size'].fillna('0', inplace=True)
df['school_size'] = df['size'].map(size_mapping)

df['school_age_years'] = 2023 - df['founded']

# Encoding sri_lankan
df['HeadQuaters'] = df['headquaters'].apply(lambda city: 1 if 'Sri Lanka' in city else 0)


# Dropping the original columns that have been encoded
df = df.drop(['Start Time', 'End Time', 'headquaters', 'size'], axis=1)

# Display the encoded DataFrame
print(df)
```

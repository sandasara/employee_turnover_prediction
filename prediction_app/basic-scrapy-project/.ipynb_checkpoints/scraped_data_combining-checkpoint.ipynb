```python
# Creating the profile list to scrape
# Check and store the past scraped data to scraped_profiles set
scraped_profiles = set()

with open('data/clean_current_and_2023_mas.json', 'r') as scraped_file:
    scraped_data = json.load(scraped_file)
    for entry in scraped_data:
        scraped_profile = entry.get("profile")
        if scraped_profile:
            scraped_profiles.add(scraped_profile)
```

```python
# List to store unique organization profile links
new_profiles = set()

# Read the excel with the complete profile list
xlsx_file = 'data/lists/profile_list.xlsx'
# Open the XLSX file
workbook = openpyxl.load_workbook(xlsx_file)
# Assuming that the profiles are in the first sheet (you can adjust accordingly)
sheet = workbook['current_mas']
# sheet = workbook['2023_from_mas']

# Read the profiles from the sheet
for row in sheet.iter_rows(min_row=2, values_only=True):
    profile = row[0]  # Assuming the profiles are in the first column
#     new_profiles.add(profile)
    if profile:
        if profile not in scraped_profiles:
            new_profiles.add(profile)

# Create the profile_list to feed to spider by converting the new_profiles set
profile_list = list(new_profiles)
len(profile_list)
```

```python
people_df = pd.read_json('data/scraped_people_data/people_2023-12-12T08-39-49.json')
people_df
```

```python
missing_exp_count = 0
missing_exp_indexes = []

# Iterate through the DataFrame
for index, row in people_df.iterrows():
    experience = row['experience']
    
    # Check if the experience list is empty
    if not experience:
        missing_exp_count += 1
        missing_exp_indexes.append(index)
        continue  # Move to the next row
    
    for item in experience:
        if item['position'] == "" or (not 'organisation' in item or item['organisation'] == ""):
            missing_exp_count += 1
            missing_exp_indexes.append(index)
            break  # No need to check further for this row

# Print the count and indexes of rows with missing experience
print("Count of rows with missing experience:", missing_exp_count)
print("Indexes of rows with missing experience:", missing_exp_indexes)
```

```python
missing_edu_count = 0
missing_edu_indexes = []

# Iterate through the DataFrame
for index, row in people_df.iterrows():
    education = row['education']
    
    # Check if the experience list is empty
    if not education:
        missing_edu_count += 1
        missing_edu_indexes.append(index)
        continue  # Move to the next row
        
    for item in education:
        if item['organisation']=="":
            missing_edu_count += 1
            missing_edu_indexes.append(index)
            break  # No need to check further for this row

# Print the count and indexes of rows with empty experience
print("Count of rows with empty education:", missing_edu_count)
print("Indexes of rows with empty education:", missing_edu_indexes)
```

```python
# Create a list with profiles that has a missing value in either experience or education
# Convert the lists to sets to remove duplicates
set1 = set(missing_exp_indexes)
set2 = set(missing_edu_indexes)

# Merge the sets using the union operator (|)
missing_set = set1 | set2

# missing_set = set1

# Convert the result set back to a list (if needed)
missing_list = list(missing_set)

len(missing_list)
```

```python
clean_profiles_df = people_df.drop(missing_list)
clean_profiles_df
```

```python
clean_profiles_df.to_json('data/cleaned_profiles/clean_profiles_14.json', orient='records')
```

```python
import json

def combine_json_files(file1_path, file2_path, output_file_path):
    # Read the contents of the first JSON file
    with open(file1_path, 'r') as file1:
        data1 = json.load(file1)

    # Read the contents of the second JSON file
    with open(file2_path, 'r') as file2:
        data2 = json.load(file2)

    # Combine the data from both files into a single list
    combined_data = data1 + data2

    # Write the combined data to a new JSON file
    with open(output_file_path, 'w') as output_file:
        json.dump(combined_data, output_file, indent=2)

# Specify the paths of your JSON files and the output file
file1_path = 'data/clean_current_and_2023_mas.json'
file2_path = 'data/cleaned_profiles/clean_profiles_14.json'
output_file_path = 'data/clean_current_and_2023_mas.json'

# Call the function to combine the JSON files
combine_json_files(file1_path, file2_path, output_file_path)

print("JSON files combined successfully.")
```

# Comapny list

```python
scraped_companies = set()

# Load the set of scraped companies from the file
with open('data/filtered_companies_combined.json', 'r') as json_file:
    json_data = json.load(json_file)
    for item in json_data:
        company_url = item.get("url")  # Get the "name" field
        if company_url:
            scraped_companies.add(company_url)  # Add the company name to the set
                       
print('Scraped: ', len(scraped_companies))
            
# List to store unique organization profile links
new_companies = set()

# Read the JSON data from the file line by line
with open('data/clean_current_and_2023_mas.json', 'r') as json_file:
    json_data = json.load(json_file)
    for line_number, profile_data in enumerate(json_data, start=1):
        try:
            # Extract the "experience" section from the JSON data
            experiences = profile_data.get("experience", [])

            # Iterate through each experience entry and extract the "organisation_profile" if it exists
            for experience in experiences:
                organization_profile = experience.get("organisation_profile")
                if organization_profile:
                    # Check if the company profile has not been scraped already
                    if organization_profile not in scraped_companies:
                        new_companies.add(organization_profile)

        except json.JSONDecodeError as e:
            print(f"Error in line {line_number}: {e}")
            print(f"Problematic line: {profile_data}")


# Convert the set to a list if needed
company_pages = list(new_companies)
print('New: ', len(company_pages))
```

```python
# split profile_list if large, to feed to the scraper
split_index = len(company_pages) // 2
company_pages1 = company_pages[:split_index]
company_pages2 = company_pages[split_index:]
```

```python
def combine_json_files(file1_path, file2_path, output_file_path):
    # Read the contents of the first JSON file
    with open(file1_path, 'r') as file1:
        data1 = json.load(file1)

    # Read the contents of the second JSON file
    with open(file2_path, 'r') as file2:
        data2 = json.load(file2)

    # Extract unique data from the second file based on the "name" attribute
    new_data = [item for item in data2 if item["name"] not in [d["name"] for d in data1]]

    # Combine the unique data with the existing data from the first file
    combined_data = data1 + new_data

    # Write the combined data to a new JSON file
    with open(output_file_path, 'w') as output_file:
        json.dump(combined_data, output_file, indent=2)

# Specify the paths of your JSON files and the output file
file1_path = 'data/filtered_companies_combined.json'
file2_path = 'data/scraped_companies_data/companies_2023-12-12T10-00-04.json'
output_file_path = 'data/filtered_companies_combined.json'

# Call the function to combine the JSON files
combine_json_files(file1_path, file2_path, output_file_path)

print("JSON files combined successfully.")
```

```python
from openpyxl import load_workbook
import json
from urllib.parse import urlencode
```

```python
scraped_profiles = set()

# Load the set of scraped profiles from the file
with open('combined_data/profiles_combined.json', 'r') as json_file:
    json_data = json.load(json_file)
    for item in json_data:
        profile = item.get("profile")  # Get the "profile" field
        if profile:
            scraped_profiles.add(profile)  # Add the profile name to the set

scraped_profiles
```

```python
# List to store unique profiles profile links
profiles_to_scrape = set()

sheet_name = 'to_scrape'
excel_file_path = 'profile_list.xlsx'

# Load the workbook
workbook = load_workbook(excel_file_path, read_only=True)
worksheet = workbook[sheet_name]

# Read data from the selected worksheet
new_profiles = {cell.value for row in worksheet.iter_rows() for cell in row}

# Close the workbook
workbook.close()

for string in new_profiles:
    if string not in scraped_profiles:
        profiles_to_scrape.add(string)
        
profiles_to_scrape
```
